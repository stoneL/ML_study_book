# 第11章 降维

```python
from scipy.stats import norm, pearsonr
from matplotlib import pylab
import numpy as np
import scipy
```

```python
def _plot_correlation_func(x, y):

    r, p = pearsonr(x, y)
    title = "Cor($X_1$, $X_2$) = %.3f" % r
    pylab.scatter(x, y)
    pylab.title(title)
    pylab.xlabel("$X_1$")
    pylab.ylabel("$X_2$")

    f1 = scipy.poly1d(scipy.polyfit(x, y, 1))
    pylab.plot(x, f1(x), "r--", linewidth=2)
    pylab.show()
```

```python
np.random.seed(0)  # to reproduce the data later on
pylab.clf()
pylab.figure(num=None, figsize=(8, 8))

x = np.arange(0, 10, 0.2)

pylab.subplot(221)
y = 0.5 * x + norm.rvs(1, scale=0.01)
_plot_correlation_func(x, y)

pylab.subplot(222)
y = 0.5 * x + norm.rvs(1, scale=0.1)
_plot_correlation_func(x, y)

pylab.subplot(223)
y = 0.5 * x + norm.rvs(1, scale=1)
_plot_correlation_func(x, y)

pylab.subplot(224)
y = norm.rvs(1, scale=10)
_plot_correlation_func(x, y)
```
